```python
#Importar librerías de Datos--------------------------------------
import numpy as np
import pandas as pd
#from scipy.stats import spearmanr

#Importar Librerías de Visualización-------------------------------
#Matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

#Seaborn
import seaborn as sns
#Dataset Seaborn
#sns.load_dataset("iris")
#sns.load_dataset("tips")
#sns.load_dataset("flights")import bokeh
#sns.set(style="darkgrid")

#Bokeh
import bokeh
#output_notebook, output_file
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.layouts import gridplot
from bokeh.models import ColumnDataSource
from bokeh.plotting import figure
```

```python
#Verificar carga de librería
output_notebook()
```

```python
#Definir valores de X
x = np.linspace(-10,10,25)
```

```python
#Definir valores de Y
y0 = [i*0.5 for i in x]
y1 = [i*2 for i in x]
y2 = [i*3 for i in x]
```

```python
source = ColumnDataSource(data=dict(x=x,y0=y0,y1=y1))
```

## Linked Panning

```python
p1= figure(tools="reset,pan", plot_width=350, plot_height=350)
p1.circle(x,y0, size= 14, color = "green", alpha=0.5)

p2 = figure(x_range=p1.x_range, y_range=p1.y_range, plot_width=350, plot_height=350)
p2.diamond(x,y1, size= 14, color = "red", alpha=0.75)

p3 = figure(x_range=p1.x_range,y_range=p1.y_range,  plot_width=350, plot_height=350)
p3.square(x,y2, size= 14, color = "blue", alpha=0.5)

p = gridplot([[p1,p2,p3]])
show(p)
```

### Deshabilitar movimiento en un eje para una gráfica

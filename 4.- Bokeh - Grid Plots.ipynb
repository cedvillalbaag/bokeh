```python
#Importar librerías
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
#Dataset Seaborn
#sns.load_dataset("iris")
#sns.load_dataset("tips")
#sns.load_dataset("flights")import bokeh
#sns.set(style="darkgrid")


#from scipy.stats import spearmanr

import bokeh
#output_notebook, output_file

from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.layouts import gridplot
from bokeh.models import ColumnDataSource
from bokeh.plotting import figure
```

```python
output_notebook()
```

```python
#Definir valores de X
x = np.linspace(-10,10,25)
```

```python
#Definir valores de Y
y0 = [i*0.5 for i in x]
y1 = [i*2 for i in x]
y2 = [i*3 for i in x]
```

```python
(x,y0),(x,y1),(x,y2)
```

## Grid Plot

```python
m0 = figure(plot_width=300,plot_height=300)
m0.circle(x,y0, size=15, color="purple", fill_color="orange")

m1 = figure(plot_width=300,plot_height=300)
m1.square(x,y1, size=15, color="purple", fill_color="pink")

m2 = figure(plot_width=300,plot_height=300)
m2.diamond(x,y2, size=15, color="purple", fill_color="red")

m= gridplot([[m0,m1,m2]]) #Atribute: Para quitar la barra usar: toolbar_location=None
show(m)
```

```python
#Importar librerías de Datos--------------------------------------
import numpy as np
import pandas as pd
#from scipy.stats import spearmanr

#Importar Librerías de Visualización-------------------------------
#Matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

#Seaborn
import seaborn as sns
#Dataset Seaborn
#sns.load_dataset("iris")
#sns.load_dataset("tips")
#sns.load_dataset("flights")import bokeh
#sns.set(style="darkgrid")

#Bokeh
import bokeh
#output_notebook, output_file
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.layouts import gridplot
from bokeh.models import ColumnDataSource
from bokeh.plotting import figure
```

```python
output_notebook()
```

```python
#Definir valores de X
x = np.linspace(-10,10,25)
```

```python
#Definir valores de Y
y0 = [i*0.5 for i in x]
y1 = [i*2 for i in x]
y2 = [i*3 for i in x]
```

```python
source = ColumnDataSource(data=dict(x=x,y0=y0,y1=y1))
```

## Linking Plots

```python
first= figure(tools="reset,pan", plot_width=350, plot_height=350)
first.circle("x","y0", size= 14, color = "red", source= source)

second= figure(tools="reset,pan", plot_width=350, plot_height=350)
second.square("x","y1", size= 14, color = "green", source= source)

final = gridplot([[first,second]])
show(final)

```

```python
first= figure(tools="reset,pan,lasso_select", plot_width=350, plot_height=350)
first.circle("x","y0", size= 14, color = "red", source= source)

second= figure(tools="reset,pan", plot_width=350, plot_height=350)
second.square("x","y1", size= 14, color = "green", source= source)

final = gridplot([[first,second]])
show(final)
```
